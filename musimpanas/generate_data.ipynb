```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(41)
data = np.random.uniform(size=(1500, 2))
data = data*4-2
data
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
medium = pd.DataFrame(data)
medium.columns = ['x', 'y']
medium['label'] = [1 if x**2+y**2 < 2.5 else 0 for x,y in data]
medium.head()
```

```python
def plot_dataset(df, ax=None):
    
    if ax is None:
        _, ax = plt.subplots(figsize=(10, 10))
        
    ax.scatter(x='x', y='y', data=df[df['label'] == 0], color='red')
    ax.scatter(x='x', y='y', data=df[df['label'] == 1], color='green')
    
    return ax
```

```python
plot_dataset(medium)
```
